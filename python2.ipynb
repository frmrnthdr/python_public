```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings('ignore')
x = np.arange(0.00, 1, 0.001)
y = np.arange(-1.0, 1.0, 0.01)
x, y = np.meshgrid(x, y)
z = y**2/x+np.log(2*np.cosh(y/x))
ax = plt.figure().gca(projection='3d')
ax.plot_surface(x, y, z, cmap=plt.cm.coolwarm)
ax.set_zlim(-50, 50)
plt.show()
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings('ignore')
x = np.arange(0.5, 5.5, 0.01)
y = np.arange(0.01, 5.01, 0.01)
x, y = np.meshgrid(x, y)
z = (y**2)/(15*x)
ax = plt.figure().gca(projection='3d')
ax.plot_surface(x, y, z, cmap=plt.cm.coolwarm)
ax.set_zlim(-0, 1)
plt.show()
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings('ignore')
J = np.arange(-5, 5, 0.01)
T = np.arange(0, 5, 0.01)
J, T = np.meshgrid(J, T)
z = 1/(J-T)
ax = plt.figure().gca(projection='3d')
ax.plot_surface(J, T, z, cmap=plt.cm.coolwarm)
ax.set_zlim(-0, 1)
plt.show()
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings('ignore')
y = np.arange(0.1, 4.1, 0.01)
x = np.arange(0.1, 8.1, 0.01)
x, y = np.meshgrid(x, y)
z = -np.log(2)+(1/2)*((x-1)/x)+(1/12)*(((x-1)/x)**2)
ax = plt.figure().gca(projection='3d')
ax.plot_surface(x, y, z, cmap=plt.cm.coolwarm)
ax.set_zlim(-1, 1)
plt.show()
```
